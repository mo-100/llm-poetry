```python
START_HEMISTICH = '<|START_HEMISTICH|>'
END_HEMISTICH = '<|END_HEMISTICH|>'
START_VERSE = '<|START_VERSE|>'
END_VERSE = '<|END_VERSE|>'
START_POEM = '<|START_POEM|>'
END_POEM = '<|END_POEM|>'
```

```python
from collections import defaultdict


# Function to format dataset for training
def format_poetry(df):
    # Group verses by poem title
    verses_dict = defaultdict(list)
    meter_dict = {}
    rhyme_dict = {}
    link_dict = {}
    for poem_title, meter, rhyme, shatr1, shatr2, link in zip(
        df["poem_title"], df['البحر'], df['قافية'], df['first_hemistich'], df['second_hemistich'], df['link']
    ):
        if meter is not None:
            meter_dict[poem_title] = meter
        if rhyme is not None:
            rhyme_dict[poem_title] = rhyme
        link_dict[poem_title] = link
        if shatr2 is None or type(shatr2) == float:
            verse = START_HEMISTICH + shatr1 + END_HEMISTICH
        else:
            verse = START_HEMISTICH + shatr1 + END_HEMISTICH + START_HEMISTICH + shatr2 + END_HEMISTICH

        verses_dict[poem_title].append(verse)

    body_dict = {}
    for t, verses in verses_dict.items():
        body_dict[t] = START_POEM + START_VERSE +  f'{END_VERSE}{START_VERSE}'.join(verses) + END_VERSE + END_POEM

    # Format the full poem in the required structure
    def get_poem(title):
        return {
            'title': title,
            'body': body_dict[title],
            'meter': meter_dict[title],
            'rhyme': rhyme_dict[title],
            'link': link_dict[title],
        }
    formatted_texts = [get_poem(t) for t in body_dict.keys()]
    return formatted_texts

df_grouped = format_poetry(df)
```

```python
nas = 0
clean_dataset = []

def is_na(item):
    kws = ['title', 'body', 'meter', 'rhyme', 'link']
    for k in kws:
        if item[k] is None or type(item[k]) == float or item[k].strip() == '':
            return True
    return False

for v in df_grouped:
    if is_na(v):
        nas += 1
    else:
        clean_dataset.append(v)
nas
```
