```python
# Load model directly
model_id = "arbml/rwkv-4-world-7b-arabic"
```

```python
import torch
from transformers import AutoTokenizer, RwkvConfig, RwkvModel

model = RwkvModel.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)

inputs = tokenizer("This is an example.", return_tensors="pt")
# Feed everything to the model
outputs = model(inputs["input_ids"])
output_whole = outputs.last_hidden_state

outputs = model(inputs["input_ids"][:, :2])
output_one = outputs.last_hidden_state

# Using the state computed on the first inputs, we will get the same output
outputs = model(inputs["input_ids"][:, 2:], state=outputs.state)
output_two = outputs.last_hidden_state

torch.allclose(torch.cat([output_one, output_two], dim=1), output_whole, atol=1e-5)
```

```python
from huggingface_hub import snapshot_download

# Download the model folder
model_dir = snapshot_download(repo_id="arbml/rwkv-4-world-7b-arabic")
vocab_dir = snapshot_download(repo_id="BlinkDL/rwkv-4-world")

# Set paths to model and vocabulary files
import os
model_path = os.path.join(model_dir, "rwkv-4-world-7b-arabic.pth")
vocab_path = os.path.join(vocab_dir, "rwkv_vocab_v20230424.json")

print(f"Model downloaded to: {model_path}")
print(f"Vocab file downloaded to: {vocab_path}")
```

```python
from rwkv.model import RWKV
from rwkv.utils import PIPELINE

# Load model
model = RWKV(model=model_dir, strategy="cpu fp32")  # Change to "cuda fp16" if using GPU
pipeline = PIPELINE(model, vocab_path)

# Generate text
prompt = "ما هو الذكاء الاصطناعي؟"
output = pipeline.generate(prompt, token_count=100)

print(output)
```
