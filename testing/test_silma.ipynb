```python
def format_poetry_prompt(poetry_text):
    """
    إنشاء توجيه محسن لإدخال الشعر إلى النموذج مع تحديد تحسين المعنى ليكون أكثر وضوحًا وترابطًا، دون تغيير الوزن أو القافية.
    """
    # إزالة الفواصل الخاصة بالشعر (مثل <|bsep|>, <|vsep|>)
    cleaned_poetry = poetry_text.replace("<|bsep|>", "").replace("</|bsep|>", "").replace("<|vsep|>", "").strip()

    # إنشاء التوجيه
    prompt = f"""
    <|system|>
    أنت نموذج لغوي متخصص في تحسين جودة الشعر العربي. دورك هو تحسين المعنى ليصبح أكثر وضوحًا وترابطًا، دون تغيير الوزن أو القافية.
    <|user|>
    أريد منك تحسين المعنى في هذا البيت الشعري مع الحفاظ على الوزن والقافية،  مع ضمان أن المعنى سيكون أكثر وضوحًا وترابطًا .
    هذا هو البيت الشعري:

    "{cleaned_poetry}"

    <|assistant|>
    النسخة المحسنة:
    """
    return prompt
```

```python
def improve_poetry(poetry_text, max_tokens=200):
    """
    تحسين معنى القصيدة مع الحفاظ على الوزن والقافية.
    """
    prompt = format_poetry_prompt(poetry_text)

    # تحويل النص إلى مدخلات النموذج
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)  # Removed .to("cuda")

    # Move inputs and model to the same device (cuda)
    inputs = inputs.to(allam_model.device)

    # توليد النص
    output = allam_model.generate(**inputs, max_new_tokens=max_tokens, do_sample=True, top_k=50, top_p=0.9, temperature=0.7)

    # استخراج النص الناتج
    improved_poetry = tokenizer.decode(output[0], skip_special_tokens=True)

    return improved_poetry
```

```python
# Step 1: Import necessary libraries
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Step 2: Load the C4AI Command R7B model and tokenizer
model_name = "CohereForAI/c4ai-command-r7b-12-2024"
tokenizer = AutoTokenizer.from_pretrained(model_name , token = hf_token)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto" , token = hf_token)

```

```python
def improve_poem(input_poem):
    prompt = f"""
تحسين القصيدة مع الحفاظ التام على الوزن والقافية

المهمة:
أنت نموذج ذكاء اصطناعي متخصص في الشعر العربي الكلاسيكي. مهمتك هي تحسين الصياغة البلاغية للقصيدة مع الحفاظ التام على الوزن والقافية دون أي تغيير في البنية الإيقاعية.

القصيدة الأصلية:
{input_poem}

قواعد صارمة يجب الالتزام بها:
- لا تغير أي كلمة تؤثر على الوزن الشعري.
- يجب أن تنتهي الأبيات بنفس القافية الأصلية.
- لا تضف أو تحذف أي كلمات تغير عدد التفاعيل في البيت.
- يمكن تحسين الأسلوب البلاغي فقط دون المساس بالبنية الإيقاعية.

تحذير:
إذا لم تحافظ القصيدة على نفس الوزن والقافية، فالتحسين غير صحيح!

طريقة التحسين:
1. خذ كل بيت كما هو ولا تغيّر عدد كلماته أو ترتيب التفعيلات.
2. استبدل بعض الكلمات فقط بأساليب أكثر بلاغة دون تغيير القافية أو الوزن.
3. لا تبدأ أي بيت بكلمة جديدة غير موجودة في الأصل.
4. اكتب القصيدة المحسنة مباشرة دون أي مقدمات أو تفسيرات.

القصيدة المحسنة:
"""



    # Tokenize the input poem with the prompt
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).to(model.device)

    generation_kwargs = {
        "max_length": 1024,  # Set a reasonable max length for the output
        "num_beams": 5,     # Beam search for quality
        "no_repeat_ngram_size": 2,  # Prevent repetition
        "temperature": 0.2,  # Control creativity
        "top_p": 0.9,       # Diversity control
        "top_k": 100,        # Limit possible next words
        "do_sample": True,  # Enable sampling (to use temperature, top_p, etc.)
    }

    # Generate the output poem with the model
    with torch.no_grad():
        output = model.generate(
            inputs.input_ids,
            attention_mask=inputs.attention_mask,  # Use attention mask to avoid issues
            pad_token_id=tokenizer.eos_token_id,   # Set padding token id
            **generation_kwargs
        )

    # Decode the output tokens to a string
    improved_poem = tokenizer.decode(output[0], skip_special_tokens=True)

    return improved_poem
```

```python
# Example input Arabic poem
input_poem = """
أَلا يا خَليليَّ ما لِعَينِكُما شَجَنا
فَما تَقطَعُ اللَيلَ مِنكُما مُتَقَطِّعُ

أَلا لَيتَ شِعري أَقوَت عَيناكُما بِما
أُحِبُّ أَم لَيسَ الَّذي أُحِبُّ مُطَّعُ

وَلَو أَنَّكُما إِذ لَجَّتِ العَينُ في الهَوى
وَلَكُما رَأَيتُما لَم تَكُما مِصطَبِعُ

"""
# Generate the improved version of the poem
improved_poem = improve_poem(input_poem)
```

```python
# Display the original and improved poem
print("Original Poem:")
print(input_poem)

print("\nImproved Poem:")
print(improved_poem)
```
